# Exploratory Data Analysis

```python
#Gender
```

```python
ax = sns.countplot(x='Gender', data =df)
```

```python
ax = sns.countplot(x='Gender', data =df)
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
df.groupby(['Gender'],as_index =False)['Amount'].sum().sort_values(by= ['Amount'],ascending = False)
```

```python
sales_gender = df.groupby(['Gender'],as_index =False)['Amount'].sum().sort_values(by= ['Amount'],ascending = False)

sns.barplot(x='Gender',y='Amount',data = sales_gender)
```

#From above graph we can say that women are spending more amount

```python
#Age
df.columns
```

```python
ax = sns.countplot(x='Age Group', data =df, hue = 'Gender')
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_age = df.groupby(['Age Group'],as_index =False)['Amount'].sum().sort_values(by= ['Amount'],ascending = False)

sns.barplot(x='Age Group',y='Amount',data = sales_age)
```

```python
# From above graphs we can say that women with 26-35 age group people spending more amount in purchasing
```

```python
#state
df.columns
```

```python
sales_state = df.groupby(['State'],as_index =False)['Orders'].sum().sort_values(by= ['Orders'],ascending = False)

sns.barplot(x='State',y='Orders',data = sales_state)
```

```python
#Total no of orders from top 10 states
sales_state = df.groupby(['State'],as_index =False)['Orders'].sum().sort_values(by= ['Orders'],ascending = False).head(10)
sns.set(rc={'figure.figsize':(20,10)})
sns.barplot(x='State',y='Orders',data = sales_state)
```

```python
#Total amount/sales from top 10 states
sales_state = df.groupby(['State'],as_index =False)['Amount'].sum().sort_values(by= ['Amount'],ascending = False).head(10)
sns.set(rc={'figure.figsize':(20,10)})
sns.barplot(x='State',y='Amount',data = sales_state)
```

```python
# From the above graphs we can say that total number of orders are more in UP, maharashtra and karntaka and 
#we can observe here keral is top 8 in amount spent but in total orders it is not in Top 10 because of less purchasing
```

```python
#Marrital_status
df.columns
```

```python
ax = sns.countplot(x='Marital_Status', data =df)
sns.set(rc={'figure.figsize':(15,10)})
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_MS = df.groupby(['Marital_Status','Gender'],as_index =False)['Amount'].sum().sort_values(by= ['Amount'],ascending = False)
sns.set(rc={'figure.figsize':(10,6)})
sns.barplot(x='Marital_Status',y='Amount',hue = 'Gender', data = sales_MS)
```

```python
#Occupation
df.columns
```

```python
ax = sns.countplot(x='Occupation', data =df)
sns.set(rc={'figure.figsize':(20,5)})
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_ocupation = df.groupby(['Occupation'],as_index =False)['Amount'].sum().sort_values(by= ['Amount'],ascending = False)
sns.set(rc={'figure.figsize':(20,10)})
sns.barplot(x='Occupation',y='Amount', data = sales_ocupation)
```

```python
# The amount who are spending are working in IT sector, Healthcare and Aviation industries
```

```python
ax = sns.countplot(x='Product_Category', data =df)
sns.set(rc={'figure.figsize':(25,10)})
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_pc = df.groupby(['Product_Category'],as_index =False)['Amount'].sum().sort_values(by= ['Amount'],ascending = False).head(10)
sns.set(rc={'figure.figsize':(20,5)})
sns.barplot(x='Product_Category',y='Amount', data = sales_pc)
```

```python
sales_pc1 = df.groupby(['Product_Category'],as_index =False)['Orders'].sum().sort_values(by= ['Orders'],ascending = False).head(10)
sns.set(rc={'figure.figsize':(20,5)})
sns.barplot(x='Product_Category',y='Orders', data = sales_pc1)
```

```python
# They spending more amount in Food, clothing and Apparel and Elecdtronics gadgets
```

```python
#Top 10 productId
sales_products = df.groupby(['Product_ID'],as_index =False)['Orders'].sum().sort_values(by= ['Orders'],ascending = False).head(10)
sns.set(rc={'figure.figsize':(20,5)})
sns.barplot(x='Product_ID',y='Orders', data = sales_products)
```
